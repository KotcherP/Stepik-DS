```
import pandas as pd
```

```
#библиотеки для визуализации. seaborn - надстройка над matplotlib. Позволяет строить более симпотичные графики и синтаксис проще
import matplotlib.pyplot as plt 
import seaborn as sns
```

```
students_performance = pd.read_csv('/content/StudentsPerformance.csv') \
    .rename(columns = {"math score" : "math_score","reading score":"reading_score","writing score":"writing_score"})
```

```
students_performance.head()
```

```
students_performance.math_score.hist() #вывести гистограмму по данным math_score средствами pandas
```

```
students_performance.plot.scatter(x='math_score',y='reading_score') #с помощью библиотеки matplotlib строится распеределение точек
```

```
#с помощью библиотеки seaborn строится распеределение точек
ax = sns.lmplot(x='math_score',y='reading_score',hue = "gender",data=students_performance,fit_reg=False)
ax.set_xlabels("Math score")
ax.set_ylabels("Reading Score") 
```
